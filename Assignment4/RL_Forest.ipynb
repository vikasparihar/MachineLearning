```python
import mdptoolbox, mdptoolbox.example
import time
import matplotlib.pyplot as plt
```

```python
P, R = mdptoolbox.example.forest()

fh = mdptoolbox.mdp.FiniteHorizon(P, R, 0.9, 3)
fh.run()

print(fh.V)
      
fh.policy
```

```python
P, R = mdptoolbox.example.forest(is_sparse=False)

pi = mdptoolbox.mdp.ValueIteration(P, R, 0.9)
pi.run()

print(pi.V)
      
print("iterations",pi.iter)
print("time",pi.time)
print("policy",pi.policy)
```

```python
P, R = mdptoolbox.example.forest(is_sparse=False)

pi = mdptoolbox.mdp.PolicyIterationModified(P, R, 0.9)
pi.run()

#print(pi.V)
print("iterations",pi.iter)
print("time",pi.time)
print("time",pi.policy)
```

```python
P, R = mdptoolbox.example.forest(is_sparse=False)

pi = mdptoolbox.mdp.QLearning(P, R, 0.1)
pi.run()

#print(pi.V)
print("Q",pi.Q)
print("time",pi.time)
print("policy",pi.policy)
```

```python
iterations =[]
time = []

for state in range(3,10000):
    P, R = mdptoolbox.example.forest(S=state, r1=10, r2=50, p=0.0, is_sparse=False)

    pi = mdptoolbox.mdp.ValueIteration(P, R, 0.9)
    pi.run()
    iterations.append(pi.iter)
    iterations.append(pi.time)

#print(pi.V)

plt.plot(range(3,10000), iterations, marker='o')
plt.xlabel('Number of States')
plt.ylabel('Numbe of Iterations')
plt.show()
    
plt.plot(range(3,10000), time, marker='o')
plt.xlabel('Number of States')
plt.ylabel('Time')
plt.show()
      
print("iterations",pi.iter)
print("time",pi.time)
print("policy",pi.policy)
```
