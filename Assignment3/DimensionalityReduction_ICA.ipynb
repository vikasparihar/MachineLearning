```python
np.random.seed =10
def ApplyICA(data):
    kurtosisAvg =[]
    reconErroAvg =[]
    timeCompute =[]
    meanerr = 0
    for i in range(1, 30):
        start_time = start_time = time.time()
        model = FastICA(n_components=i,max_iter=100000,tol=0.01)        
        data_new = model.fit_transform(data)
        timeCompute.append((time.time()-start_time) *1000)
        df = pd.DataFrame(data_new)
        kurt_avg = df.kurtosis().mean()       
        kurtosisAvg.append(kurt_avg)
        df_inv=model.inverse_transform(df)
        if(type(data).__name__ == 'DataFrame'):
            meanerr = (df_inv-data).apply(np.absolute).mean().mean()
        else:
            meanerr = np.absolute((df_inv-data)).mean().mean()
        reconErroAvg.append(meanerr)
    plt.plot(range(1, 30), kurtosisAvg, marker='o')
    plt.xlabel('Number of components')
    plt.ylabel('Avg Kurtosis Score')
    plt.show()
    
    plt.plot(range(1, 30), reconErroAvg, marker='o')
    plt.xlabel('Number of components')
    plt.ylabel('Avg Reconstruction Error')
    plt.show()
    
    plt.plot(range(1, 30), timeCompute, marker='o')
    plt.xlabel('Number of components')
    plt.ylabel('Time (in milliseconds)')
    plt.show()
   
```
